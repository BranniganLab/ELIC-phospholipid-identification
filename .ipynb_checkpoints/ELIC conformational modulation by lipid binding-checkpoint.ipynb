```python
%matplotlib widget
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
from scipy import constants
```

```python
#Changes in free energy relative to PC
DG_PG_5  = -7
DG_PG_WT = -11
DG_PE_5  = -2
DG_PE_WT = -9
```

```python
#linspace
N = 1000 #number of subsamples
xmin = 1e-12 
xmax = .9

grid_x = np.linspace(xmin, xmax, N)

xPG, xPE = np.meshgrid(grid_x, grid_x) #titration grid

xPC =  np.full((N, N), 1.0) - (xPG + xPE) #PC+PG+PE=1
mask = xPC<0
xPG=np.where(~mask, xPG, np.nan)
xPE=np.where(~mask, xPE, np.nan)
```

```python
#logspace
N = 1000 #number of subsamples
xmin = -12 
xmax = -0.1

grid_x = np.logspace(xmin, xmax, N)

xPG, xPE = np.meshgrid(grid_x, grid_x) #titration grid

xPC =  np.full((N, N), 1.0) - (xPG + xPE) #PC+PG+PE=1
mask = xPC<0
xPG=np.where(~mask, xPG, np.nan)
xPE=np.where(~mask, xPE, np.nan)
```

```python
temperature = 303.15 #Kelvin
RT = temperature*constants.R/(1000*constants.calorie) #kcal/mol

#Boltzmann weights
KCG5  = np.exp(-DG_PG_5 / RT)
KCGWT = np.exp(-DG_PG_WT / RT)
KCE5  = np.exp(-DG_PE_5 / RT)
KCEWT = np.exp(-DG_PE_WT / RT)

data = (xPC + KCG5 * xPG + KCE5 * xPE) / (xPC + KCGWT * xPG + KCEWT * xPE)
data = np.log(data)


#data=np.where(mask, data, np.nan)
```

```python
%matplotlib widget

plt.plot(np.log10(xPG), np.log10(xPE), ".k", markersize=1)
plt.contourf(np.log10(xPG), np.log10(xPE), data, levels=100)

plt.xlabel("log PG content")
plt.ylabel("log PE content")
plt.title("Relative log-probability of ELIC5 conformation")

plt.colorbar()
plt.savefig('logspace_conformationBias.png', dpi=600)
```

```python
%matplotlib widget
#plt.contourf(np.log(xPG), np.log(xPE), data, levels=100)
plt.plot(xPG, xPE, ".k")
plt.contourf(xPG, xPE, data, levels=100)
# plt.contour(np.log10(grid_x), np.log10(grid_y), data, levels=40, color='black')
plt.xlabel("log PG content")
plt.ylabel("log PE content")
plt.title("Relative log-probability of ELIC5 conformation")

plt.xscale("log")
plt.yscale("log") 
plt.xlim([1e-12,1])
plt.ylim([1e-12,1])
plt.colorbar()
plt.savefig('linspace_conformationBias.png', dpi=600)
```

```python
%matplotlib widget
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(np.log10(xPG), np.log10(xPE), data)
ax.set_xlabel('log PG')
ax.set_ylabel('log PE')
ax.set_zlabel('log(E5) conformation')

ax.set_title('3D contour')
plt.show()
```
