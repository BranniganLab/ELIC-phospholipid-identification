```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("step7.2.dat", delim_whitespace=True, names = ["step", "bond", "angle", "dihed", "elect", "vdw", "boundary", "kinetic", "total", "temp", "potential", "tempavg", "pressure", "volume"])
```

```python
data = data.set_index("step")
```

```python
data.volume.plot()
var = "Volume"
plt.title(f"{var} vs step (step size = 2fs)")
plt.ylabel(var)
```

```python
data.pressure.plot()
var = "Pressure"
plt.title(f"{var} vs step (step size = 2fs)")
plt.ylabel(var)
```

```python
data.total.plot()
var = "Total Energy"
plt.title(f"{var} vs step (step size = 2fs)")
plt.ylabel(var)
```
