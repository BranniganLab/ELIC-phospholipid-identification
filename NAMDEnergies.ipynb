```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy
```

```python
def plotEnergies (data, timeConst, prefix, suffix):
    fig, (time, dist) = plt.subplots(1,2)

    time.plot(data.Time/timeConst, data.Total)
    time.set_xlabel(f"Time in ns")
    time.set_ylabel("Nonbonded Energy (kcal/mol)")
    
    vals, edges = np.histogram(data.Total, density=True, bins=20)
    X = (edges[:-1]+edges[1:])/2
    dist.plot(X, vals)
    dist.set_xlabel("Nonbonded Energy (kcal/mol)")
    dist.set_ylabel("Density")
    
    fig.suptitle(f"{prefix}\n{suffix}\nMean: {np.round(np.mean(data.Total),1)} +/- {np.round(np.std(data.Total), 1)} kcal/mol")
    fig.tight_layout()
    fig.set_figwidth(15)
```

```python
WTPG = pd.read_csv("./WT_POPC/ARG318_POEG", delim_whitespace=True)
plotEnergies(WTPG, 1000, "ARG318 and PG", "Coupling: 0.75 to 1")
plt.savefig("WTPG.png", dpi=600)
```

```python
WTPE = pd.read_csv("./WT_POPC/ARG318_PE", delim_whitespace=True)
plotEnergies(WTPE, 10000, "ARG318 and PE", "Decoupling L0 to L0.25")
plt.savefig("WTPE.png", dpi=600)
```
